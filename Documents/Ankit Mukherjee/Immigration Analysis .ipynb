```python
# Dependencies 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Files to load
immigration_file_to_load = (r"C:\Users\bayek\OneDrive\Documents\GitHub\Project-1-DeskTop\canadian_immegration_data.csv")
home_ownership_to_load = (r"C:\Users\bayek\OneDrive\Documents\GitHub\Project-1-DeskTop\to-buy-or-to-rent.csv")
ownership_rate_to_load = (r"C:\Users\bayek\OneDrive\Documents\GitHub\Project-1-DeskTop\home-ownership-rate.csv")
```

```python
# Read the files and study the results 
immigration_metadata = pd.read_csv(immigration_file_to_load)
home_ownership = pd.read_csv(home_ownership_to_load)
ownership_rate = pd.read_csv(ownership_rate_to_load)
```

```python
# Display the Immigration CSV
immigration_metadata.head()
```

```python
# Set the country name as index
immigration_metadata.set_index('Country', inplace=True,)
immigration_metadata.head()
```

```python
# Convert columns 1980 to 2013 to years 
years = list(map(str, range(1980, 2013)))
```

```python
# Sort the Countries by Total in descending order
top_countries = immigration_metadata.sort_values(['Total'], ascending=False,)

# Get only the top five entries
top_five = top_countries.head()

# Transpose the data frame
top_five = top_five[years].transpose()

top_five.head()
```

```python
# Compare the trends of top 5 countries that contributed the most to immigration to Canada using an area graph.
top_five.index = top_five.index.map(int)
top_five.plot(kind='area',
              stacked=False,
             figsize=(20, 10))  # pass a tuple (x, y) size

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# Sort the Countries by Total in Ascending order
bottom_countries = immigration_metadata.sort_values(['Total'], ascending=True,)

# Get only the top five entries
bottom_five = bottom_countries.head()

# Transpose the data frame
bottom_five = bottom_five[years].transpose()

bottom_five.head()
```
