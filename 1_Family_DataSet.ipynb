```python
labels = Dualearnersdf['Year']
values1 = Dualearnersdf['VALUE']
values2 =singleearnersdf['VALUE']

# set width of bars
barWidth = 0.6
# adjust spacing between bars
plt.subplots_adjust(wspace=0.3)

# set positions of bars on x-axis
r1 = np.arange(len(values1))
r2 = [x + barWidth for x in r1]

# create bar graph
plt.bar(r1, values1, color='blue', width=barWidth, edgecolor='black', label='Dual Earners')
plt.bar(r2, values2, color='red', width=barWidth, edgecolor='black', label='Single Earners')

# set axis labels and legend
plt.xlabel('Year')
plt.ylabel('Number')
plt.xticks([r + barWidth/2 for r in range(len(values1))], labels, rotation=90)
plt.legend()
plt.savefig(r"C:\Users\Eleshia\Desktop\DA Boot Camp\Renters-Anonymous\Images\Types of Earning Families.png")
```
